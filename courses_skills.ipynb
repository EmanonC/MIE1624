```
import pandas as pd
```

```
data = pd.read_csv('skills.csv', index_col=0)
```

```
data = data.sort_values('count')
data = data.reset_index(drop=True)
data
```

```
courses = pd.read_excel('courses.xlsx', index_col=0)
courses.columns = ['courses name', 'topic', 'prer']
courses = courses.reset_index()
courses.drop(columns='index', inplace=True)
courses.drop(columns='prer', inplace=True)
```

```
def remove(shit):
  if 'Game-playing AI with Swift for TensorFlow' in shit:
    return 'Game-playing AI with Swift for TensorFlow'
  else:
    return shit
courses['courses name'] = courses['courses name'].apply(remove)
courses['courses name'].value_counts()
```

Convert to lower case

```
def toLower(shit):
  return shit.lower()

courses['courses name'] = courses['courses name'].apply(toLower)
courses['topic'] = courses['topic'].apply(toLower)
```

Search in course names

```
result = []
for i in range(len(data['skill'])):
  try:
    result += list(set(courses[courses['courses name'].str.contains(data['skill'][i])]['courses name']))*i
  except:
    pass
```

```
from collections import Counter
fuck = pd.Series(Counter(result))
```

```
fuck = pd.DataFrame(fuck)
fuck = fuck.reset_index()
```

```
fuck.columns = ['course name', 'score']
fuck = fuck.sort_values('score', ascending=False)
fuck
```
